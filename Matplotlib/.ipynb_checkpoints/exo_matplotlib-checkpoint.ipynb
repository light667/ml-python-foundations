EXERCICE: 
Vous avez un dictionnaire qui comprend N datasets, creez une fonction qui tra√ße N graphiques sur une seule figure

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = {f"experience{i}": np.random.randn(100, 2) for i in range(12)}
```

```python
def graphique(dataset):
    n = len(dataset)
    plt.figure(figsize= (12, 30))
    for i, j in zip (dataset.keys(), range(1, n+1)):
        plt.subplot(n, 1, j)
        plt.plot(dataset[i])
        plt.title(i)
    plt.show()
```

```python
graphique(dataset)
```
