```python
import numpy as np

x = np.linspace(0, 2, 10)
y = x**2
print(x)
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y)
plt.show()
```

```python
# pour tracer un nuage de point :
plt.scatter(x, y)
```

```python
# Customiser la courbe
# nom de la courbe : label
# l"epaisseur de la courbe : lw
# Style du trait : ls
# Couleur du trait : C
```

```python
plt.plot(x, y, c='black', lw= 5, ls='--')
plt.show()
```

```python
"""
Cycle de vie d'une courbe
"""
```

```python
# on commence par :
plt.figure()  # a crer une feuille de travail
plt.plot(x, y, label = 'quadratique')
plt.plot(x, x**3, label= 'cubique')
plt.title('figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.show()
plt.savefig('figure.png')
```

Pour generer plusieur graphique sur une seule figure : on utilise la methode subplot

```python
plt.subplot(2, 1, 1)
plt.plot(x, y, c='red')
plt.subplot(2, 1, 2)
plt.plot(x, np.sin(x), label= 'sin')
plt.plot(x, np.cos(x), label= 'cos')
plt.legend()
```

Partie 2

Utilisation de la methode OOP pour la construction des graphiques

```python
fig , ax = plt.subplots(2, 1, sharex= True)
ax[0].plot(x, y)
ax[1].plot(x, np.sin(x))
plt.show()
```
