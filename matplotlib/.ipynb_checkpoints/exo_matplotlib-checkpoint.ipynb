Vous avez un dictionnaire qui comprend N datasets, creez une fonction qui tra√ße N graphiques sur une seule figure

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = {f"experience{i}": np.random.randn(100) for i in range(4)}
```

```python
def graphique(dataset):
    for i in range(4):
        plt.subplot(4, 1, i+1)
        plt.plot(np.linspace(0, 100, 100), dataset[f"experience{i}"])
        plt.show()
```

```python
graphique(dataset)
```
